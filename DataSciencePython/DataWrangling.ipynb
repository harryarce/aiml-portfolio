```python
# data wrangling 
#missing values
df = pd.read_csv('class-grades.csv')
df.describe()
```

```python
df.isna().any()
```

```python
# mean imputation
from sklearn.impute import SimpleImputer
import numpy as np
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
mean_imputer = mean_imputer.fit(df)
imputed_df = mean_imputer.transform(df.values)
df1 = pd.DataFrame(data=imputed_df, columns=['Prefix','Assignment', 'Tutorial', 'Midterm', 'TakeHome','Final'])
print(df1.isna().any())
print(df1.head(9))
```

```python
# mean imputation
from sklearn.impute import SimpleImputer
import numpy as np
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
mean_imputer = mean_imputer.fit(df)
imputed_df = mean_imputer.transform(df.values)
df1 = pd.DataFrame(data=imputed_df, columns=['Prefix','Assignment', 'Tutorial', 'Midterm', 'TakeHome','Final'])
df1.head(9)
```

```python
# outliers
import seaborn as sns
sns.boxplot(x=df1['Assignment'])
```

```python
#outlier treatment
filter=df1['Assignment'].values>60
df1_outlier_rem=df1[filter]
df1_outlier_rem
```

```python
# groupby
import pandas as pd 
world_cup={ 'Team' : [ 'West Indies' , 'West indies' ,'India', 'Australia','Pakistan' ,'Sri Lanka','Australia', 'Australia', 'Australia', 'Insia', 'Australia'], 
            'Rank' : [7,7,2,1,6,4,1,1,1,2,1] , 
            'Year' : [1975, 1979,1983, 1987, 1992, 1996, 1999, 2003, 2007, 2011, 2015]}
df=pd. DataFrame (world_cup) 
print (df.groupby ( [ 'Team' , 'Rank'] ).groups) 
```

```python
# concatenation   add without relationship

import pandas as pd
world_champions = {'Team': ['lndia', 'Australia', 'West Indies', 'Pakistan', 'Sri Lanka'],
                   'ICC_rank': [2, 3, 7, 8, 4],
                   'world_champions_year': [2011, 2015, 1979, 1992, 1996],
                   'Points': [874, 787, 753, 673, 855]}
chokers = {'Team': ['South Africa', 'New Zealand', 'Zimbabwe'],
           'ICC_rank': [1, 5, 9], 
           'Points': [874, 787, 753]}
dfl = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print(dfl)
print(df2)
print(pd.concat([dfl, df2], axis=1))
print(pd.concat([dfl, df2], axis=0))
```

```python
# join types
#left join
print (pd. merge (dfl, df2, on= 'Team' , how= 'left' ) ) 
#right
print (pd. merge (dfl, df2, on= 'Team' , how= 'right' ) ) 
#inner join
print (pd. merge (dfl, df2, on= 'Team' , how= 'inner' ) ) 
#full outer
print (pd. merge (dfl, df2, on= 'Team' , how= 'outer' ) ) 
```
