# Exploratory Data Analysis

```python
# exploratory data analysis EDA
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
boston_ds = load_boston()
```

```python
x_axis = boston_ds.data
```

```python
y_axis = boston_ds.target
```

```python
style.use('ggplot')
plt.figure(figsize=(10,10))
plt.hist(y_axis, bins=50)
plt.xlabel('price in 1000s USD')
plt.ylabel('number of houses')
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(10,10))
plt.scatter(boston_ds.data[:,5],boston_ds.target)
plt.xlabel('price in 1000s USD')
plt.ylabel('number of houses')
plt.show()
```

```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df_auto = pd.read_csv('auto-mpg.csv')
```

```python
df_auto.head()
```

```python
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Asia'

df_auto['origin'] = df_auto['origin'].apply(origin)
```

```python
df_auto.head(30)
```

```python
sns.pairplot(df_auto[['mpg','weight','origin']],hue='origin',height=4)
```
